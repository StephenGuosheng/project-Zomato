```python
# import all the required libraries
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter
import numpy as np
# use the inline backend
%matplotlib inline
```

```python
# import dataset istanbul
url = 'https://drive.google.com/file/d/1U3bzevmt2Gh4s4HKTyTpK_6bIwa2dGVl/view?usp=sharing' # dataset download link
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2] # get dataset file id
pd_istanbul = pd.read_csv(url,encoding = 'utf-8',index_col=0)
pd_istanbul
```

```python
# import dataset seattle
url = 'https://drive.google.com/file/d/1Rrf0M_Ru59bCKrWVGnGPY4swVlyE8xUL/view?usp=sharing' # dataset download link
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2] # get dataset file id
pd_seattle = pd.read_csv(url,encoding='ISO-8859-1', index_col=0)
pd_seattle
```

```python
# dataset singapore
url = 'https://drive.google.com/file/d/12_8a-vPHw7U3PeVpJ2D1E88uBg6I1g4S/view?usp=sharing' # dataset download link
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2] # get dataset file id
pd_singapore = pd.read_csv(url,encoding='ISO-8859-1', index_col=0)
pd_singapore
```

```python
# The genre of food with highest average rating in three cities
# istanbul
df1 = pd_istanbul.drop(pd_istanbul.index[pd_istanbul['rating'] == '-'], inplace = False)
df1['rating'] = df1['rating'].astype(float)
df1 = df1.groupby('genre').mean()

df1=df1.sort_values(by='rating', ascending=False)
df1
```

```python
# seattle
df2 = pd_seattle.drop(pd_seattle.index[pd_seattle['rating'] == '-'], inplace = False)
df2['rating'] = df2['rating'].astype(float)
df2 = df2.groupby('genre').mean()

df2=df2.sort_values(by='rating', ascending=False)
df2
```

```python
# Singapore
df3 = pd_singapore.drop(pd_singapore.index[pd_singapore['rating'] == '-'], inplace = False)
df3['rating'] = df3['rating'].astype(float)
df3 = df3.groupby('genre').mean()

df3= df3.sort_values(by='rating', ascending=False)
df3
```

```python
df1=df1[:10]
df2=df2[:10]
df3=df3[:10]
df3
```

```python
# apply the ggplot style
mpl.style.use('ggplot') # optional: for ggplot-like style
```

```python
# create a plots for the three cities, listing the top 10 most-popular (in terms of rating) food in each of the three cities

fig = plt.figure(figsize=(25,15)) # create figure

ax0 = fig.add_subplot(1, 3, 1) # add subplot 1 (1 row, 3 columns, first plot)
ax1 = fig.add_subplot(1, 3, 2) # add subplot 2 (1 row, 3 columns, second plot)
ax2 = fig.add_subplot(1, 3, 3) # as above

# Subplot 1: 
sns.barplot(x=df1['rating'],y=df1.index, palette='deep',ax=ax0)
ax0.set_title ('Instanbul',y=1.05)
ax0.set_xlabel('Rating')
ax0.set_yticklabels(ax0.get_yticklabels(),rotation=70)

# Subplot 2: 
sns.barplot( x=df2['rating'],y=df2.index, palette='deep',ax=ax1)
ax1.set_title ('Seattle',y=1.05)
ax1.set_xlabel('Rating')
ax1.set_yticklabels(ax1.get_yticklabels(),rotation=70)

# Supplot 3:
sns.barplot( x=df3['rating'],y= df3.index, palette='deep',ax=ax2)
ax2.set_title ('Singapore',y=1.05)
ax2.set_xlabel('Rating')
ax2.set_yticklabels(ax2.get_yticklabels(),rotation=70)

plt.suptitle('Top 10 highest-rating food in three cities',fontsize = 28)
plt.show()
```

```python
# to see the distribution of the cateen rating in three different cities
# extract the data to be compared, i.e., the ratings of each city

dfr_istanbul = pd_istanbul.drop(pd_istanbul.index[pd_istanbul['rating'] == '-'], inplace = False)
dfr_istanbul['rating'] = dfr_istanbul['rating'].astype(float)
data1=dfr_istanbul['rating']

dfr_seattle = pd_seattle.drop(pd_seattle.index[pd_seattle['rating'] == '-'], inplace = False)
dfr_seattle['rating'] = dfr_seattle['rating'].astype(float)
data2=dfr_seattle['rating']

dfr_singapore = pd_singapore.drop(pd_singapore.index[pd_singapore['rating'] == '-'], inplace = False)
dfr_singapore['rating'] = dfr_singapore['rating'].astype(float)
data3=dfr_singapore['rating']
```

```python
# make three histograms out of the dataframe
# color to be changed

plt.figure(figsize=(18,8))
bins = [x+n for n in range(1, 5) for x in [0.0, 1]]+[5.0]

arr=plt.hist(data1, bins, alpha=0.1, color = 'orangered', label='Istanbul')
for i in range(8):
  if arr [0][i]>0:
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]),ha='left')

arr=plt.hist(data2, bins, alpha=0.1, color = 'lightskyblue', label='Seattle')
for i in range(8):
  if arr [0][i]>0:
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]),ha='left')

arr=plt.hist(data3, bins, alpha=0.1, color = 'violet',label='Singapore')
for i in range(8):
  if arr [0][i]>0:
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]),ha='left')

plt.xticks(range (5))
plt.legend(loc='upper left', fontsize=18)
plt.title('The rating distribution in three cities',fontsize=28,y=1.2)
plt.ylabel('number of canteens')
plt.xlabel('Ratings')

plt.show()
```

```python
# Compare the price distribution of the three cities

dfp_istanbul= pd_istanbul.assign(price = lambda x: x['price'].str.extract('(\d+)'))
dfp_istanbul['price'] = dfp_istanbul['price'].astype(int)
dfp_istanbul
```

```python
dfp_seattle= pd_seattle.assign(price = lambda x: x['price'].str.extract('(\d+)'))
dfp_seattle = dfp_seattle.dropna(subset= ['price'])
dfp_seattle['price'] = dfp_seattle['price'].astype(int)
dfp_seattle
```

```python
dfp_singapore = pd_singapore.assign(price = lambda x: x['price'].str.extract('(\d+)'))
dfp_singapore = dfp_singapore.dropna(subset= ['price'])
dfp_singapore['price'] = dfp_singapore['price'].astype(int)
dfp_singapore
```

```python
# make three histogram
fig = plt.figure(figsize=(20,10)) # create figure

ax0 = fig.add_subplot(1, 3, 1) # add subplot 1 (1 row, 3 columns, first plot)
ax1 = fig.add_subplot(1, 3, 2) # add subplot 2 (1 row, 3 columns, second plot)
ax2 = fig.add_subplot(1, 3, 3) # as above

#Istanbul
ax0.hist (dfp_istanbul['price'], bins=10, color = 'orangered', edgecolor='white',weights=np.ones_like(dfp_istanbul['price'])*100 / len(dfp_istanbul['price']))
ax0.yaxis.set_major_formatter(PercentFormatter())
ax0.set_title('Istanbul ', y=1.12)
ax0.set_xlabel('price (in Turkish lira for two people)')

# Seattle
ax1.hist(dfp_seattle['price'], bins=10, color = 'lightskyblue', edgecolor='white',weights=np.ones_like(dfp_seattle['price'])*100 / len(dfp_seattle['price']))
ax1.yaxis.set_major_formatter(PercentFormatter())
ax1.set_title('Seattle ', y=1.12)
ax1.set_xlabel('price (in USD for two people)')

# Singapore
ax2.hist(dfp_singapore['price'], bins=10, color ='violet',edgecolor='white',weights=np.ones_like(dfp_singapore['price'])*100 / len(dfp_singapore['price']))
ax2.yaxis.set_major_formatter(PercentFormatter())
ax2.set_title('Singapore ', y=1.12)
ax2.set_xlabel('price (in SGD for two people)')


plt.suptitle('The distribution of price for canteens in three cities', y=1.1, fontsize=28)

plt.show()
```

```python
# the distribution of price in boxplot

fig = plt.figure(figsize=(20,13)) # create figure

ax0 = fig.add_subplot(1, 3, 1) # add subplot 1 (1 row, 3 columns, first plot)
ax1 = fig.add_subplot(1, 3, 2) # add subplot 2 (1 row, 3 columns, second plot)
ax2 = fig.add_subplot(1, 3, 3) # as above

# Istanbul
sns.boxplot(y='price', data=dfp_istanbul, color = 'orangered',ax=ax0)
ax0.set_title('Istanbul ', y=1.1)
ax0.set_ylabel('price (in Turkish lira for two people)')

# Seattle
sns.boxplot(y='price', data=dfp_seattle,color = 'lightskyblue',ax=ax1)
ax1.set_title('Seattle ', y=1.1)
ax1.set_ylabel('price (in USD for two people)')

# Singapore

sns.boxplot(y='price', data=dfp_singapore, color ='violet',ax=ax2)
ax2.set_title('Singapore', y=1.1)
ax2.set_ylabel('price (in SGD for two people)')

plt.suptitle('The distribution of price for canteens in three cities',y=1.05, fontsize=28)

plt.show()
```

```python
# see relationship between rating and price (does higher price mean higher rating? )
# convert the rating in each table from string to int

dfr_istanbul = dfp_istanbul.drop(dfp_istanbul.index[dfp_istanbul['rating'] == '-'], inplace = False)
dfr_istanbul['rating'] = dfr_istanbul['rating'].astype(float)

dfr_seattle = dfp_seattle.drop(dfp_seattle.index[dfp_seattle['rating'] == '-'], inplace = False)
dfr_seattle['rating'] = dfr_seattle['rating'].astype(float)

dfr_singapore = dfp_singapore.drop(dfp_singapore.index[dfp_singapore['rating'] == '-'], inplace = False)
dfr_singapore['rating'] = dfr_singapore['rating'].astype(float)
```

```python
# make a jointplot of the three cities to see the relationship
fig = plt.figure(figsize=(25,10)) # create figure

ax0 = fig.add_subplot(1, 3, 1) # add subplot 1 (1 row, 3 columns, first plot)
ax1 = fig.add_subplot(1, 3, 2) # add subplot 2 (1 row, 3 columns, second plot)
ax2 = fig.add_subplot(1, 3, 3) # as above

# Istanbul
dfr_istanbul.plot(kind='scatter',color = 'orangered', x='price', y='rating', ax=ax0)
ax0.set_title('Istanbul',y=1.05)
ax0.set_ylabel('rating)')
ax0.set_xlabel('price (in Turkish lira for two people)')

# Seattle
dfr_seattle.plot(kind='scatter',color = 'lightskyblue', x='price', y='rating', ax=ax1)
ax1.set_title('Seattle', y=1.05)
ax1.set_ylabel('rating)')
ax1.set_xlabel('price (in USD for two people)')

# Singapore
dfr_singapore.plot(kind='scatter',color ='violet', x='price', y='rating', ax=ax2)
ax2.set_title('Singapore',y=1.05)
ax2.set_ylabel('rating)')
ax2.set_xlabel('price (in USD for two people)')

plt.suptitle('Canteen distribution in terms of price and rating in three cities',y=1.05, fontsize=28)

plt.show()
```

```python
# to check the R^2 of price and rating of the three cities
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
#Istanbul
x0= dfr_istanbul[['price']]
y0= dfr_istanbul.rating
model.fit(x0,y0)
Rs1= model.score(x0,y0)

#Seattle
x1= dfr_seattle[['price']]
y1= dfr_seattle.rating
model.fit(x1,y1)
Rs2= model.score(x1,y1)

#Singapore
x2= dfr_singapore[['price']]
y2= dfr_singapore.rating
model.fit(x2,y2)
Rs3= model.score(x2,y2)

print ('the R square for Istanbul canteen price and rating is',Rs1)
print ('the R square for Seattle canteen price and rating is',Rs2)
print ('the R square for Singapore canteen price and rating is',Rs3)
```

Obviously, cateen price could explain **little** of its rating
