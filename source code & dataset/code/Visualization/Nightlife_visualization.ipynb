```python
# import all the required libraries
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter
import numpy as np
# use the inline backend
%matplotlib inline
```

```python
# import dataset seattle
url = 'https://drive.google.com/file/d/1dxPhpTCsweuGHhQbUdzkVzGZOJFuPLIg/view?usp=sharing' # dataset download link
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2] # get dataset file id
pd_seattle = pd.read_csv(url,encoding='ISO-8859-1',index_col=0)
pd_seattle
```

```python
# import dataset rome
url = 'https://drive.google.com/file/d/1zSeSwsYL2S_cu2BXsn0f3RbImL_jB4Q0/view?usp=sharing' # dataset download link
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2] # get dataset file id
pd_rome = pd.read_csv(url,encoding = 'utf-8',index_col=0)
pd_rome
```

```python
# import dataset mumbai
url = 'https://drive.google.com/file/d/1P8ne2hENHwwoD0Eys4SnaJxqf8y2M9B9/view?usp=sharing' # dataset download link
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2] # get dataset file id
pd_mumbai = pd.read_csv(url,encoding = 'utf-8',index_col=0)
pd_mumbai
```

```python
# The genre of food with highest average rating in three cities
# seattle
df1 = pd_seattle.drop(pd_seattle.index[pd_seattle['rating'] == '-'], inplace = False)
df1['rating'] = df1['rating'].astype(float)
df1 = df1.groupby('genre').mean()

df1=df1.sort_values(by='rating', ascending=False)
df1
```

```python
df2 = pd_rome.drop(pd_rome.index[pd_rome['rating'] == '-'], inplace = False)
df2['rating'] = df2['rating'].astype(float)
df2 = df2.groupby('genre').mean()

df2=df2.sort_values(by='rating', ascending=False)
df2
```

```python
# mumbai
df3 = pd_mumbai.drop(pd_mumbai.index[pd_mumbai['rating'] == '-'], inplace = False)
df3['rating'] = df3['rating'].astype(float)
df3 = df3.groupby('genre').mean()

df3= df3.sort_values(by='rating', ascending=False)
df3
```

```python
df1=df1[:10]
df2=df2[:10]
df3=df3[:10]
df3
```

```python
# apply the ggplot style
mpl.style.use('ggplot') # optional: for ggplot-like style
```

```python
# create a plots for the three cities, listing the top 10 most-popular (in terms of rating) food in each of the three cities

fig = plt.figure(figsize=(25,15)) # create figure

ax0 = fig.add_subplot(1, 3, 1) # add subplot 1 (1 row, 3 columns, first plot)
ax1 = fig.add_subplot(1, 3, 2) # add subplot 2 (1 row, 3 columns, second plot)
ax2 = fig.add_subplot(1, 3, 3) # as above

# Subplot 1: 
sns.barplot(x=df1['rating'],y=df1.index, palette='deep',ax=ax0)
ax0.set_title ('seattle',y=1.05)
ax0.set_xlabel('Rating')
ax0.set_yticklabels(ax0.get_yticklabels(),rotation=70)

# Subplot 2: 
sns.barplot( x=df2['rating'],y=df2.index, palette='deep',ax=ax1)
ax1.set_title ('rome',y=1.05)
ax1.set_xlabel('Rating')
ax1.set_yticklabels(ax1.get_yticklabels(),rotation=70)

# Supplot 3:
sns.barplot( x=df3['rating'],y= df3.index, palette='deep',ax=ax2)
ax2.set_title ('mumbai',y=1.05)
ax2.set_xlabel('Rating')
ax2.set_yticklabels(ax2.get_yticklabels(),rotation=70)

plt.suptitle('Top 10 most popular food for nightlife in three cities',fontsize = 28)
plt.show()
```

```python
# to see the distribution of the cateen rating in three different cities
# extract the data to be compared, i.e., the ratings of each city

dfr_seattle = pd_seattle.drop(pd_seattle.index[pd_seattle['rating'] == '-'], inplace = False)
dfr_seattle['rating'] = dfr_seattle['rating'].astype(float)
data1=dfr_seattle['rating']

dfr_rome = pd_rome.drop(pd_rome.index[pd_rome['rating'] == '-'], inplace = False)
dfr_rome['rating'] = dfr_rome['rating'].astype(float)
data2=dfr_rome['rating']

dfr_mumbai = pd_mumbai.drop(pd_mumbai.index[pd_mumbai['rating'] == '-'], inplace = False)
dfr_mumbai['rating'] = dfr_mumbai['rating'].astype(float)
data3=dfr_mumbai['rating']
```

```python
# make three histograms out of the dataframe
# color to be changed

plt.figure(figsize=(18,8))
bins = [x+n for n in range(1, 5) for x in [0.0, 1]]+[5.0]

arr=plt.hist(data1, bins, alpha=0.1, color = 'orangered', label='seattle')
for i in range(8):
  if arr [0][i]>0:
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]),ha='left')

arr=plt.hist(data2, bins, alpha=0.1, color = 'lightskyblue', label='rome')
for i in range(8):
  if arr [0][i]>0:
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]),ha='left')

arr=plt.hist(data3, bins, alpha=0.1, color = 'violet',label='mumbai')
for i in range(8):
  if arr [0][i]>0:
    plt.text(arr[1][i],arr[0][i],str(arr[0][i]),ha='left')

plt.xticks(range (5))
plt.legend(loc='upper left', fontsize=18)
plt.title('The rating distribution in three cities',fontsize=28,y=1.2)
plt.ylabel('number of canteens')
plt.xlabel('Ratings')

plt.show()
```

```python
# Compare the price distribution of the three cities

dfp_seattle= pd_seattle.assign(price = lambda x: x['price'].str.extract('(\d+)'))
dfp_seattle = dfp_seattle.dropna(subset= ['price'])
dfp_seattle['price'] = dfp_seattle['price'].astype(int)
dfp_seattle
```

```python
dfp_rome= pd_rome.assign(price = lambda x: x['price'].str.extract('(\d+)'))
dfp_rome['price'] = dfp_rome['price'].astype(int)
dfp_rome
```

```python
dfp_mumbai= pd_mumbai.assign(price = lambda x: x['price'].str.extract('(\d+)'))
dfp_mumbai['price'] = dfp_mumbai['price'].astype(int)
dfp_mumbai
```

```python
# make three histogram
fig = plt.figure(figsize=(20,10)) # create figure

ax0 = fig.add_subplot(1, 3, 1) # add subplot 1 (1 row, 3 columns, first plot)
ax1 = fig.add_subplot(1, 3, 2) # add subplot 2 (1 row, 3 columns, second plot)
ax2 = fig.add_subplot(1, 3, 3) # as above

#seattle
ax0.hist (dfp_seattle['price'], bins=10, color = 'orangered', edgecolor='white',weights=np.ones_like(dfp_seattle['price'])*100 / len(dfp_seattle['price']))
ax0.yaxis.set_major_formatter(PercentFormatter())
ax0.set_title('seattle ', y=1.12)
ax0.set_xlabel('price (in USD for two people)')

# rome
ax1.hist(dfp_rome['price'], bins=10, color = 'lightskyblue', edgecolor='white',weights=np.ones_like(dfp_rome['price'])*100 / len(dfp_rome['price']))
ax1.yaxis.set_major_formatter(PercentFormatter())
ax1.set_title('rome ', y=1.12)
ax1.set_xlabel('price (in euro for two people)')

# mumbai
ax2.hist(dfp_mumbai['price'], bins=10, color ='violet',edgecolor='white',weights=np.ones_like(dfp_mumbai['price'])*100 / len(dfp_mumbai['price']))
ax2.yaxis.set_major_formatter(PercentFormatter())
ax2.set_title('mumbai ', y=1.12)
ax2.set_xlabel('price (in INR for two people)')


plt.suptitle('The distribution of price for restaurants in three cities', y=1.1, fontsize=28)

plt.show()
```

```python
# the distribution of price in boxplot

fig = plt.figure(figsize=(20,13)) # create figure

ax0 = fig.add_subplot(1, 3, 1) # add subplot 1 (1 row, 3 columns, first plot)
ax1 = fig.add_subplot(1, 3, 2) # add subplot 2 (1 row, 3 columns, second plot)
ax2 = fig.add_subplot(1, 3, 3) # as above

# seattle
sns.boxplot(y='price', data=dfp_seattle, color = 'orangered',ax=ax0)
ax0.set_title('seattle ', y=1.1)
ax0.set_ylabel('price (in USD for two people)')

# rome
sns.boxplot(y='price', data=dfp_rome,color = 'lightskyblue',ax=ax1)
ax1.set_title('rome ', y=1.1)
ax1.set_ylabel('price (in euro for two people)')

# mumbai

sns.boxplot(y='price', data=dfp_mumbai, color ='violet',ax=ax2)
ax2.set_title('mumbai', y=1.1)
ax2.set_ylabel('price (in INR for two people)')

plt.suptitle('The distribution of price for canteens in three cities',y=1.05, fontsize=28)

plt.show()
```

```python
# see relationship between rating and price (does higher price mean higher rating? )
# convert the rating in each table from string to int

dfr_seattle = dfp_seattle.drop(dfp_seattle.index[dfp_seattle['rating'] == '-'], inplace = False)
dfr_seattle['rating'] = dfr_seattle['rating'].astype(float)

dfr_rome = dfp_rome.drop(dfp_rome.index[dfp_rome['rating'] == '-'], inplace = False)
dfr_rome['rating'] = dfr_rome['rating'].astype(float)

dfr_mumbai = dfp_mumbai.drop(dfp_mumbai.index[dfp_mumbai['rating'] == '-'], inplace = False)
dfr_mumbai['rating'] = dfr_mumbai['rating'].astype(float)
```

```python
# make a jointplot of the three cities to see the relationship
fig = plt.figure(figsize=(25,10)) # create figure

ax0 = fig.add_subplot(1, 3, 1) # add subplot 1 (1 row, 3 columns, first plot)
ax1 = fig.add_subplot(1, 3, 2) # add subplot 2 (1 row, 3 columns, second plot)
ax2 = fig.add_subplot(1, 3, 3) # as above

# seattle
dfr_seattle.plot(kind='scatter',color = 'orangered', x='price', y='rating', ax=ax0)
ax0.set_title('seattle',y=1.05)
ax0.set_ylabel('rating)')
ax0.set_xlabel('price (in USD for two people)')

# rome
dfr_rome.plot(kind='scatter',color = 'lightskyblue', x='price', y='rating', ax=ax1)
ax1.set_title('rome', y=1.05)
ax1.set_ylabel('rating)')
ax1.set_xlabel('price (in euro for two people)')

# mumbai
dfr_mumbai.plot(kind='scatter',color ='violet', x='price', y='rating', ax=ax2)
ax2.set_title('mumbai',y=1.05)
ax2.set_ylabel('rating)')
ax2.set_xlabel('price (in INR for two people)')

plt.suptitle('Restaurants distribution in terms of price and rating in three cities',y=1.05, fontsize=28)

plt.show()
```

```python
# to check the R^2 of price and rating of the three cities
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
#seattle
x0= dfr_seattle[['price']]
y0= dfr_seattle.rating
model.fit(x0,y0)
Rs1= model.score(x0,y0)

#rome
x1= dfr_rome[['price']]
y1= dfr_rome.rating
model.fit(x1,y1)
Rs2= model.score(x1,y1)

#mumbai
x2= dfr_mumbai[['price']]
y2= dfr_mumbai.rating
model.fit(x2,y2)
Rs3= model.score(x2,y2)

print ('the R square for seattle canteen price and rating is',Rs1)
print ('the R square for rome canteen price and rating is',Rs2)
print ('the R square for mumbai canteen price and rating is',Rs3)
```
