```python
# import all the required libraries
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter
import numpy as np
# use the inline backend
%matplotlib inline
```

```python
# import dataset istanbul
url = 'https://drive.google.com/file/d/1lET10htZzY-N2r2uCISGq-aByRl6nRo1/view?usp=sharing' # dataset download link
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2] # get dataset file id
pd_is = pd.read_csv(url,encoding = 'utf-8',index_col=0)
pd_is
```

```python
# import dataset seattle
url = 'https://drive.google.com/file/d/1u2YiTEZXeB6-X5Y67Ou82FMnTyFN-Kb9/view?usp=sharing' # dataset download link
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2] # get dataset file id
pd_sea = pd.read_csv(url,encoding='ISO-8859-1', index_col=0)
pd_sea
```

```python
# import dataset lisbon
url = 'https://drive.google.com/file/d/1xGWNg_aMkHaM8Dx8ZL2wuF9kLrun83ao/view?usp=sharing' # dataset download link
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2] # get dataset file id
pd_lis = pd.read_csv(url,encoding='ISO-8859-1', index_col=0)
pd_lis
```

```python
# see the sum to thematic places in three cities
pd_istanbul = pd_is
pd_seattle = pd_sea
pd_lisbon = pd_lis

pd_istanbul['City']='Istanbul'
df1=pd_istanbul
pd_seattle['City']='Seattle'
df2=pd_seattle
pd_lisbon['City']='Lisbon'
df3=pd_lisbon
```

```python
# concatenate three dfs into one
df_sum = pd.concat([df1, df2, df3], axis=0)
df_sum
```

```python
# see which cities has the most places for fun
dfs =df_sum.rename(columns = {'number of places':'places'}, inplace = False)
dfs = dfs.assign (places = lambda x: x['places'].str.extract('(\d+)'))
dfs ['places'] = dfs['places'].astype(int)
dfs1 = dfs.reset_index(drop=True, inplace=False)
```

```python
# calculate the sum of places
dfs2 = dfs1.groupby(['City'], axis=0).sum()
dfs2
```

```python
dfs3 = dfs1.groupby(['City'], axis=0).count()
dfs3
```

```python
# to a bar chart 
dfs3 = dfs3['theme']
df_comp = pd.concat([dfs2, dfs3], axis=1)
df_comp = df_comp.sort_values(by='places', ascending=False)
```

```python
df_comp
```

```python
# apply the ggplot style
mpl.style.use('ggplot') # optional: for ggplot-like style
```

```python
# to a bar chart
df_comp.plot( kind= 'bar' , figsize=(20, 12), secondary_y='theme',rot=5) #inplace = true
plt.title ('The overall number of themes and realted places in three cities',fontsize = 28, y=1.1)

plt.show()
```

```python
# top 10 most popular themes in three cities

# Istanbul
dfi =pd_is.rename(columns = {'number of places':'places'}, inplace = False)
dfi = dfi.assign (places = lambda x: x['places'].str.extract('(\d+)'))
dfi ['places'] = dfi['places'].astype(int)
dfi=dfi.sort_values(by='places', ascending=False)

# Seattle
dfsea =pd_sea.rename(columns = {'number of places':'places'}, inplace = False)
dfsea = dfsea.assign (places = lambda x: x['places'].str.extract('(\d+)'))
dfsea ['places'] = dfsea['places'].astype(int)
dfsea=dfsea.sort_values(by='places', ascending=False)

# Lisbon
dflis =pd_lis.rename(columns = {'number of places':'places'}, inplace = False)
dflis = dflis.assign (places = lambda x: x['places'].str.extract('(\d+)'))
dflis ['places'] = dflis['places'].astype(int)
dflis=dflis.sort_values(by='places', ascending=False)

dfi=dfi[:10]
dfsea=dfsea[:10]
dflis=dflis[:10]
```

```python
df_dis1 =  dfi[['theme', 'places']]
df_dis1 = df_dis1.set_index('theme')
df_dis2 =  dfsea[['theme', 'places']]
df_dis2 = df_dis2.set_index('theme')
df_dis3 =  dflis[['theme', 'places']]
df_dis3 = df_dis3.set_index('theme')
```

```python
df_dis11 = df_dis1.sort_values(by = 'places', ascending=True, inplace=False)
df_dis22 = df_dis2.sort_values(by = 'places', ascending=True, inplace=False)
df_dis33 = df_dis3.sort_values(by = 'places', ascending=True, inplace=False)
```

```python
# Istanbul
df_dis11.plot(kind='barh',color='orangered',figsize=(20, 10))
plt.title ('The distribution of themes in terms of number in Istanbul',y=1.05,fontsize = 28)
plt.xlabel('number of theme places')
plt.yticks(rotation = 60)

plt.show()
```

```python
# draw the pie charts to see the distribution of food in these cities

fig = plt.figure(figsize=(25, 10)) # create figure

ax0 = fig.add_subplot(1, 3, 1) # add subplot 1 (1 row, 3 columns, first plot)
ax1 = fig.add_subplot(1, 3, 2) # add subplot 2 (1 row, 3 columns, second plot)
ax2 = fig.add_subplot(1, 3, 3) # as above

colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink','linen','mediumorchid','olive','thistle']

# Subplot 1: pie chart
df_dis1.plot(kind='pie',autopct='%1.1f%%',pctdistance=1.12, colors = colors_list, labels=None,ax=ax0,subplots=True) # add to subplot 1
ax0.set_title('Istanbul ', y=1.2) 
# add legend
ax0.legend(labels=df_dis1.index, bbox_to_anchor=(1.2,1.3), loc='upper right')

# Subplot 2: pie chart
df_dis2.plot(kind='pie',autopct='%1.1f%%', pctdistance=1.12,colors = colors_list, labels=None, ax=ax1,subplots=True) # add to subplot 2
ax1.set_title('Seattle ', y=1.2)
ax1.legend(labels=df_dis2.index, bbox_to_anchor=(1.2,1.3), loc='upper right')


# Subplot 2: pie chart
df_dis3.plot(kind='pie',autopct='%1.1f%%', pctdistance=1.12, colors = colors_list, labels=None, ax=ax2,subplots=True) # add to subplot 2
ax2.set_title('Lisbon ', y=1.2)
ax2.legend(labels=df_dis3.index, bbox_to_anchor=(1.2,1.3), loc='upper right')

plt.suptitle('The distribution of Top 10 most popular themes in three cities', y=1.05, fontsize=28)
```
